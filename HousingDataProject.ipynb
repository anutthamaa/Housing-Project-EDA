##### Observation:
   1. Heatmap is plotted for the correlation matrix obtained above.
   2. Threshold value is set to 0.6 or -0.6
   3. Variables correlation value is to be compared with threshold value.

```python
#Filtering top 10 variables which are highly correlated with SalePrice
k = 10
columns = numeric_housing_df.corr().nlargest(k, 'SalePrice')['SalePrice'].index
corr_matrix = numeric_housing_df[columns].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot = True, annot_kws= {'size' : 10},
           cmap = "coolwarm", fmt = '.2f', linewidths = 2, linecolor = 'white')
plt.show()
```

##### Observation:
   1. 'OverallQual','GrLivArea','GarageArea','TotalBmstSF' are highly correlated with SalePrice.
   2. 'GarageCars' and 'GarageArea' are highly correlated. So, no.of cars that can fit depends on the garage area built.As it is highly correlated we can drop one of these columns.
   3. TotalBsmtSF is highly correlated with 1stFlrSF. So, we can remove one of these.
   4. GrLivArea is highly correlated with TotRmsAbvGrd. So, we can remove one of these.

```python
#pearsonr method returns the p value and correlation coefficient value for all numeric variables against sale price
from scipy.stats import pearsonr

significant_columns = []
for column in numeric_housing_df.columns:
    if(column != 'SalePrice'):
        corr,p = pearsonr(numeric_housing_df.SalePrice, numeric_housing_df[column])
        if((corr >= 0.5) | (corr <= -0.5)):
            print(column , corr, p)
            significant_columns.append(column)
```

##### Observation:
   1. The threshold value has been set between >=0.5 and <=-0.5.
   2. Total 10 variables are correlated with the target variable SalePrice.

```python
#Function to plot based on the plot type.
def Plot(df, plottype):
    ix = 1
    fig = plt.figure(figsize = (15,10))
    for c in list(df.columns):
        if ix <= 3:
            ax2 = fig.add_subplot(2,3,ix+3)
            ax2.tick_params(axis='x', labelrotation=90)
            if(plottype == 'count'):
                sns.countplot(data = df, x = c, ax = ax2) #for countplot   
            elif(plottype == 'box'):
                sns.boxplot(x=df[c], y=numeric_housing_df.SalePrice, ax=ax2) #for boxplot    
            elif(plottype == 'outliers'):
                sns.boxplot(y=df[c], ax=ax2) #boxplot for outliers     
            fig.tight_layout()
        ix = ix +1
        if ix == 4: 
            fig = plt.figure(figsize = (15,10))
            ix = 1
```

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency

class ChiSquare:
#Function to determine p-value and perform chi-square test
    def __init__(self, dataframe):
        self.df = dataframe
        self.p = None #P-Value
        self.chi2 = None #Chi-square Test Statistic
        self.dof = None
        
        self.dfObserved = None
        self.dfExpected = None

        
#Function to determine chi-square and p-value less than or equal to 0.05
    def TestIndependence(self,columns, alpha=0.05):      
        list1 = []
        list2 = []
        
        for colX in columns:
            
            X = self.df[colX].astype(str)
            Y = numeric_housing_df['SalePrice'].astype(str)
        
            self.dfObserved = pd.crosstab(Y,X) 
            chi2, p, dof, expected = stats.chi2_contingency(self.dfObserved.values)
            self.p = p
            self.chi2 = chi2
            self.dof = dof 
        
            self.dfExpected = pd.DataFrame(expected, columns=self.dfObserved.columns, index = self.dfObserved.index)
            
            if self.p<alpha:
                list1.append(colX)
            else:
                list2.append(colX)
            
       
        print('Categorical variables IMPORTANT for Prediction are :',list1)
        print()
        print('Categorical variables NOT IMPORTANT for Prediction are:', list2 ,  '\n(Hence these variables can be DISCARDED from model)')
        

#Initializing ChiSquare Class
cT = ChiSquare(categorical_housing_df)

#Perform Feature Selection
testColumns = categorical_housing_df.columns

cT.TestIndependence(columns = testColumns) 
```
